```python
def play_game():
    score = 0

    print("\n=== THE LOST TREASURE ADVENTURE ===")
    name = input("Enter your name: ")
    print(f"\nWelcome, {name}! Your mission is to find the lost treasure.")

    # First decision
    print("\nYou arrive at a crossroads.")
    print("1. Go into the forest")
    print("2. Walk towards the mountains")

    choice1 = input("Choose (1/2): ")

    if choice1 == "1":
        score += 10
        print("\nYou enter the forest and find a mysterious cave.")

        print("\n1. Enter the cave")
        print("2. Continue through the forest")
        choice2 = input("Choose (1/2): ")

        if choice2 == "1":
            score += 20
            print("\nInside the cave, you discover a treasure chest!")

            print("\n1. Open the chest")
            print("2. Leave it alone")
            choice3 = input("Choose (1/2): ")

            if choice3 == "1":
                score += 30
                ending = "Treasure Hunter Ending"
                print("\nYou found the legendary treasure!")
            else:
                score += 10
                ending = "Wise Explorer Ending"
                print("\nYou leave safely, avoiding possible traps.")

        else:
            score += 15
            ending = "Forest Survivor Ending"
            print("\nYou safely leave the forest with valuable herbs.")

    elif choice1 == "2":
        score += 10
        print("\nYou head toward the mountains.")

        print("\n1. Climb the steep path")
        print("2. Take the safer route")
        choice2 = input("Choose (1/2): ")

        if choice2 == "1":
            score += 25
            ending = "Mountain Hero Ending"
            print("\nYou reach the summit and discover ancient ruins.")
        else:
            score += 15
            ending = "Traveler Ending"
            print("\nYou enjoy a peaceful journey and return home.")

    else:
        ending = "Lost Ending"
        print("\nInvalid choice. You got lost!")

    print("\n===== GAME OVER =====")
    print("Player:", name)
    print("Ending:", ending)
    print("Final Score:", score)

# Replay loop
while True:
    play_game()

    replay = input("\nDo you want to play again? (yes/no): ").lower()

    if replay != "yes":
        print("\nThanks for playing!")
        break
```
